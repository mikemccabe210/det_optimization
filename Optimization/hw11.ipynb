```python
from scipy.linalg import svd

clown = np.loadtxt('clownImage.txt')
U, S, V = svd(clown)
fig, axes = plt.subplots(3, 1, figsize = (12, 6))
dims = [5, 15, 25]
for i, ax in enumerate(axes):
    k = dims[i]
    V_k = V[:k, :]
    S_k = np.diag(S[:k])
    U_k = U[:, :k]
    k_clown = U_k.dot(S_k.dot(V_k))
    a = ax.imshow( k_clown)
    ax.axis('off')
    ax.set_title(str(k))
plt.tight_layout()
```
